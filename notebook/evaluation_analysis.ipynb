```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import joblib
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, 
    f1_score, roc_auc_score, confusion_matrix, 
    roc_curve, classification_report, ConfusionMatrixDisplay
)
import warnings
warnings.filterwarnings('ignore') 
%matplotlib inline 

# Create results directory 
import os  
os.makedirs('../results', exist_ok=True)  

print(" Libraries imported successfully")
# ## 2. Load Data and Models
print("="*70)
print("STEP 1: LOADING PROCESSED DATA")
print("="*70)

# Load the processed data
data_path = '../data/processed_diabetes_data.pkl'
data = joblib.load(data_path)

print("Data loaded successfully")
print(f"Data type: {type(data)}")
print(f"Keys in data dictionary: {list(data.keys())}")

# Extract the data splits
X_train = data['X_train']
y_train = data['y_train']
X_val = data['X_val']      
y_val = data['y_val']
X_test = data['X_test']   
y_test = data['y_test']

print(f"\nðŸ“Š DATA SPLITS:")
print(f"X_train: {X_train.shape} - Training features")
print(f"y_train: {y_train.shape} - Training labels")
print(f"X_val:   {X_val.shape}   - Validation features (for evaluation)")
print(f"y_val:   {y_val.shape}   - Validation labels")
print(f"X_test:  {X_test.shape}  - Test features (holdout set)")
print(f"y_test:  {y_test.shape}  - Test labels")
if hasattr(y_val, 'values'):
    y_val = y_val.values
if hasattr(y_test, 'values'):
    y_test = y_test.values

print(f"\n CLASS DISTRIBUTION IN VALIDATION SET:")
unique, counts = np.unique(y_val, return_counts=True)
for cls, cnt in zip(unique, counts):
    percentage = (cnt / len(y_val)) * 100
    label = "No Diabetes (0)" if cls == 0 else "Diabetes (1)"
    print(f"  {label}: {cnt} samples ({percentage:.1f}%)")
print("\n" + "="*70)
print("STEP 2: LOADING TRAINED MODELS")
print("="*70)

models = {}
model_files = {
    'Logistic Regression': '../models/logistic_regression_model.pkl',
    'Random Forest': '../models/random_forest_model.pkl',
    'SVM': '../models/svm_model.pkl',
    'XGBoost': '../models/xgboost_model.pkl'
}

for name, path in model_files.items():
    try:
        model = joblib.load(path)
        models[name] = model
        print(f" {name:25} -> Loaded successfully")
        print(f"   Model type: {model.__class__.__name__}")
        
        if hasattr(model, 'get_params'):
            params = model.get_params()
            key_params = {k: params[k] for k in list(params.keys())[:3]}  
            print(f"   Key params: {key_params}")
            
    except Exception as e:
        print(f" {name:25} -> Failed to load: {type(e).__name__}")

print(f"\n Total models loaded: {len(models)}")
print("\n" + "="*70)
print("STEP 3: MAKING PREDICTIONS")
print("="*70)
predictions = {}
probabilities = {}

for name, model in models.items():
    try:
        # Make predictions
        y_pred = model.predict(X_val)
        predictions[name] = y_pred
        
        # Get probabilities for ROC-AUC
        if hasattr(model, 'predict_proba'):
            y_prob = model.predict_proba(X_val)[:, 1]  
            probabilities[name] = y_prob
            print(f" {name:25} -> Predictions + Probabilities")
        else:
            probabilities[name] = None
            print(f"  {name:25} -> Predictions only (no probabilities)")
            
    except Exception as e:
        print(f" {name:25} -> Error: {type(e).__name__}")

print(f"\n Predictions made for {len(predictions)} models")
print("\n" + "="*70)
print("STEP 4: CALCULATING EVALUATION METRICS")
print("="*70)

results = []

for name in models.keys():
    y_pred = predictions.get(name)
    
    if y_pred is not None:
        try:
            # Calculate metrics
            acc = accuracy_score(y_val, y_pred)
            prec = precision_score(y_val, y_pred, zero_division=0)
            rec = recall_score(y_val, y_pred, zero_division=0)
            f1 = f1_score(y_val, y_pred, zero_division=0)
            
            # ROC-AUC (requires probabilities)
            y_prob = probabilities.get(name)
            if y_prob is not None:
                roc_auc = roc_auc_score(y_val, y_prob)
            else:
                roc_auc = None
            
            results.append({
                'Model': name,
                'Accuracy': round(acc, 4),
                'Precision': round(prec, 4),
                'Recall': round(rec, 4),
                'F1-Score': round(f1, 4),
                'ROC-AUC': round(roc_auc, 4) if roc_auc is not None else 'N/A'
            })
            
            print(f" {name:25} -> Metrics calculated")
            
        except Exception as e:
            print(f" {name:25} -> Error calculating metrics: {e}")
    else:
        print(f" {name:25} -> No predictions available")

# Create results DataFrame
results_df = pd.DataFrame(results)

print("\n" + "="*70)
print("PERFORMANCE COMPARISON TABLE")
print("="*70)
print(results_df.to_string(index=False))

# Save to CSV
results_df.to_csv('../results/model_performance.csv', index=False)
print(f"\n Performance table saved to: ../results/model_performance.csv")
print("\n" + "="*70)
print("STEP 5: GENERATING CONFUSION MATRICES")
print("="*70)

# Create 2x2 grid for confusion matrices
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
axes = axes.ravel()

# Get list of model names
model_names = list(models.keys())

for idx, name in enumerate(model_names[:4]):  
    y_pred = predictions.get(name)
    
    if y_pred is not None:
        # Calculate confusion matrix
        cm = confusion_matrix(y_val, y_pred)
        
        # Plot confusion matrix
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx],
                   xticklabels=['Predicted No', 'Predicted Yes'],
                   yticklabels=['Actual No', 'Actual Yes'],
                   cbar=False)
        
        axes[idx].set_title(f'{name}\nConfusion Matrix', fontsize=14, fontweight='bold')
        axes[idx].set_xlabel('Predicted Label', fontsize=12)
        axes[idx].set_ylabel('Actual Label', fontsize=12)
        
        # Calculate and display accuracy
        accuracy = np.trace(cm) / np.sum(cm)
        axes[idx].text(0.5, -0.15, f'Accuracy: {accuracy:.3f}', 
                      ha='center', transform=axes[idx].transAxes, fontsize=11,
                      bbox=dict(boxstyle='round,pad=0.3', facecolor='lightblue', alpha=0.5))
        
        # Add TP, FP, FN, TN annotations
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
            axes[idx].text(0.5, -0.25, f'TN={tn}, FP={fp}, FN={fn}, TP={tp}', 
                          ha='center', transform=axes[idx].transAxes, fontsize=10)

plt.tight_layout()
plt.savefig('../results/confusion_matrices.png', dpi=300, bbox_inches='tight')
plt.show()
print(" Confusion matrices saved to: ../results/confusion_matrices.png")
print("\n" + "="*70)
print("STEP 6: GENERATING ROC CURVES")
print("="*70)

plt.figure(figsize=(10, 8))
models_with_probs = []
for name in models.keys():
    y_prob = probabilities.get(name)
    
    if y_prob is not None:
        try:
            fpr, tpr, _ = roc_curve(y_val, y_prob)
            auc = roc_auc_score(y_val, y_prob)
            
            plt.plot(fpr, tpr, linewidth=2.5, label=f'{name} (AUC = {auc:.3f})')
            models_with_probs.append(name)
        except:
            pass

if models_with_probs:
    plt.plot([0, 1], [0, 1], 'k--', linewidth=1.5, alpha=0.7, 
            label='Random Classifier (AUC = 0.500)')
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.title('ROC Curves - Model Comparison', fontsize=16, fontweight='bold')
    plt.legend(loc='lower right', fontsize=11)
    plt.grid(True, alpha=0.3)
    plt.xlim([-0.01, 1.01])
    plt.ylim([-0.01, 1.01])
    plt.text(0.6, 0.3, 'AUC Interpretation:', fontsize=11, fontweight='bold')
    plt.text(0.6, 0.25, '0.9-1.0: Excellent', fontsize=10)
    plt.text(0.6, 0.20, '0.8-0.9: Very Good', fontsize=10)
    plt.text(0.6, 0.15, '0.7-0.8: Good', fontsize=10)
    plt.text(0.6, 0.10, '0.6-0.7: Fair', fontsize=10)
    plt.text(0.6, 0.05, '0.5-0.6: Poor', fontsize=10)
    
    plt.savefig('../results/roc_curves.png', dpi=300, bbox_inches='tight')
    plt.show()
    print(f" ROC curves saved to: ../results/roc_curves.png")
    print(f"   Models included: {', '.join(models_with_probs)}")
else:
    print("  No models have probability scores for ROC curves")
print("\n" + "="*70)
print("STEP 7: DETAILED CLASSIFICATION REPORTS")
print("="*70)

for name in models.keys():
    y_pred = predictions.get(name)
    
    if y_pred is not None:
        print(f"\n{'='*60}")
        print(f"MODEL: {name}")
        print(f"{'='*60}")
        report = classification_report(y_val, y_pred, 
                                      target_names=['No Diabetes', 'Diabetes'],
                                      digits=4)
        print(report)
        cm = confusion_matrix(y_val, y_pred)
        if cm.shape == (2, 2):
            tn, fp, fn, tp = cm.ravel()
            
            print(" Confusion Matrix Details:")
            print(f"  True Negatives (TN):  {tn:3d} - Correctly predicted non-diabetic")
            print(f"  False Positives (FP): {fp:3d} - Non-diabetic predicted as diabetic (Type I error)")
            print(f"  False Negatives (FN): {fn:3d} - Diabetic predicted as non-diabetic (Type II error)")
            print(f"  True Positives (TP):  {tp:3d} - Correctly predicted diabetic")
            specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
            fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
            fnr = fn / (fn + tp) if (fn + tp) > 0 else 0
            
            print(f"\n   Additional Metrics:")
            print(f"  Specificity (True Negative Rate): {specificity:.4f}")
            print(f"  False Positive Rate: {fpr:.4f}")
            print(f"  False Negative Rate: {fnr:.4f}")
print("\n" + "="*70)
print("STEP 8: MODEL RANKING AND ANALYSIS")
print("="*70)

if not results_df.empty:
    print("\n PERFORMANCE RANKINGS:")
    print("-" * 50)
    metrics_to_rank = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
    
    for metric in metrics_to_rank:
        if metric in results_df.columns:
            valid_df = results_df[results_df[metric] != 'N/A']
            if not valid_df.empty:
                ranked = valid_df.sort_values(by=metric, ascending=False)
                
                print(f"\n **{metric} Ranking:**")
                for i, (_, row) in enumerate(ranked.iterrows(), 1):
                    medal = ""
                    if i == 1: medal = ""
                    elif i == 2: medal = ""
                    elif i == 3: medal = ""
                    
                    value = row[metric]
                    if isinstance(value, (int, float)):
                        print(f"   {medal} {i:2d}. {row['Model']:25} = {value:.4f}")
                    else:
                        print(f"   {medal} {i:2d}. {row['Model']:25} = {value}")
    
    print("\n" + "-" * 50)
    print(" CLINICAL INTERPRETATION FOR DIABETES PREDICTION:")
    print("-" * 50)
    print("""
    â€¢ **Accuracy**: Overall correctness rate
      - Good for balanced datasets
    
    â€¢ **Precision** (Positive Predictive Value):
      - High = Few false positives
      - Important when unnecessary treatments are costly/harmful
    
    â€¢ **Recall** (Sensitivity):
      - High = Few false negatives  
      - CRITICAL for diabetes - missing cases is dangerous!
    
    â€¢ **F1-Score**:
      - Harmonic mean of Precision and Recall
      - Best for imbalanced medical data
    
    â€¢ **ROC-AUC**:
      - Overall discriminative ability
      - AUC > 0.8 = Good, > 0.9 = Excellent
    """)
print("\n" + "="*70)
print("STEP 9: SAVING COMPREHENSIVE REPORT")
print("="*70)
report_content = f"""============================================================
COMPREHENSIVE MODEL EVALUATION REPORT
============================================================

EVALUATION SUMMARY
----------------------------------------
Evaluation Date          : {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}
Total Models Evaluated   : {len(models)}
Validation Set Size      : {X_val.shape[0]} samples
Test Set Size (holdout)  : {X_test.shape[0]} samples
Target Variable          : Diabetes (Binary: 0=No Diabetes, 1=Diabetes)

DATA SPLIT INFORMATION
----------------------------------------
Training Set   : {X_train.shape[0]} samples
Validation Set : {X_val.shape[0]} samples (used for evaluation)
Test Set       : {X_test.shape[0]} samples (holdout for final validation)

Class Distribution (Validation Set):
  No Diabetes (0): {np.sum(y_val == 0)} samples ({np.sum(y_val == 0)/len(y_val)*100:.1f}%)
  Diabetes (1)   : {np.sum(y_val == 1)} samples ({np.sum(y_val == 1)/len(y_val)*100:.1f}%)

============================================================
PERFORMANCE METRICS
============================================================
{results_df.to_string(index=False)}

============================================================
KEY FINDINGS
============================================================
1. **Best Overall Model**: 
   [Will be determined from the results above]

2. **Clinical Recommendations**:
   â€¢ For maximum detection (avoiding missed cases): Choose model with highest Recall
   â€¢ For minimizing false alarms: Choose model with highest Precision
   â€¢ For balanced performance: Choose model with highest F1-Score
   â€¢ For overall discriminative ability: Choose model with highest ROC-AUC

3. **Confusion Matrix Analysis**:
   â€¢ True Positives: Correctly identified diabetes cases
   â€¢ False Negatives: Missed diabetes cases (clinically critical)
   â€¢ False Positives: False alarms (lead to unnecessary tests)
   â€¢ True Negatives: Correctly identified non-diabetes cases

============================================================
RECOMMENDATIONS
============================================================
1. For clinical use where false negatives are critical (missing diabetes cases):
   - Prioritize models with HIGH RECALL

2. For screening where follow-up tests are expensive:
   - Prioritize models with HIGH PRECISION

3. For general purpose/balanced approach:
   - Choose model with HIGHEST F1-SCORE

4. For overall discriminative ability:
   - Choose model with HIGHEST ROC-AUC

============================================================
LIMITATIONS & FUTURE WORK
============================================================
1. **Dataset Limitations**:
   - Limited sample size
   - Class imbalance in diabetes data
   - Features may not capture all diabetes indicators

2. **Model Limitations**:
   - All models struggle with borderline cases
   - Feature engineering could improve performance
   - External validation needed for clinical deployment

3. **Future Improvements**:
   - Collect more diverse patient data
   - Include additional clinical features
   - Try ensemble methods
   - Implement cost-sensitive learning for medical context
"""
with open('../results/comprehensive_evaluation_report.txt', 'w') as f:
    f.write(report_content)

print(" Comprehensive report saved to: ../results/comprehensive_evaluation_report.txt")


print("\n" + "="*70)
print(" EVALUATION COMPLETED SUCCESSFULLY!")
print("="*70)
print("\n **Generated Files in 'results/' folder:**")
print("1. model_performance.csv          - Performance metrics table")
print("2. confusion_matrices.png         - Confusion matrix visualizations")
print("3. roc_curves.png                 - ROC curve comparison")
print("4. comprehensive_evaluation_report.txt - Complete analysis report")
print("5. model_evaluation_summary.xlsx  - Excel summary with interpretations")
```
